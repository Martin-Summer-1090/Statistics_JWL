### Exercise 3: Where on the line are you?

Line through (2,1) with slope 1. Note: You can sketch this by hand. I have done this here in R just for my convenience to put the plot into the notebook.

```r
plot.new()

grid(nx = 20, ny = 20,
    lty = 2, col = "gray", lwd = 2)

par(new = TRUE)

curve(-1 + x, from = -10, to = 10, lwd = 2, col = "red")
abline(v = 0)
abline(h = 0)
```

Slope -1

```r
plot.new()

grid(nx = 20, ny = 20,
    lty = 2, col = "gray", lwd = 2)

par(new = TRUE)

curve(3 - x, from = -10, to = 10, lwd = 2, col = "red")
abline(v = 0)
abline(h = 0)
```

### Exercise 1: Life expectancy

```r
curve(67.78 + 0.0002549*x, from = 0, to = 10000,
     main = "Relation of GDP per capita and life expectancy",
     xlab = "GDP per capita in international Dollars at 2011 prices",
     ylab = "Average life expectancy at birth in years"    
     )
```

```r
library(JWL)
lex  <- life_expectancy
```

```r
head(lex)
```

```r
lex_2018  <- lex[lex$Year == 2018, ]
```

```r
head(lex_2018)
```

```r
sel  <- c("Chad", "Egypt", "Kenya", "Vietnam", "Thailand", "India", "Switzerland", "United States", "Qatar")
```

```r
countries  <- lex_2018[lex_2018$Entity %in% sel, ] 
```

```r
data  <- countries[ ,c("Entity", "GDPC", "LE")]
```

```r
data
```

```r
curve(67.78 + 0.0002549*x, from = 0, to = 160000,
     main = "Relation of GDP per capita and life expectancy",
     xlab = "GDP per capita in international Dollars at 2011 prices",
     ylab = "Average life expectancy at birth in years"    
     )

points(data$GDPC, data$LE, pch = 19, col = "red")
text(x = data$GDPC + 2, y = data$LE + 2, label = data$Entity)
```

Add my own country
